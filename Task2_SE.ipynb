```python
def calculate_grade(m1, m2, m3):
    avg = (m1 + m2 + m3) / 3
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= 40:
        return "D"
    else:
        return "F"

print(calculate_grade(70, 80, 90))
```

```python
def grade(avg):
    if avg >= 90: return "A"
    elif avg >= 75: return "B"
    elif avg >= 60: return "C"
    elif avg >= 40: return "D"
    else: return "F"

m1, m2, m3 = map(int, input("Enter 3 marks: ").split())
avg = (m1 + m2 + m3) / 3
print("Grade:", grade(avg))
```

```python
class Student:
    def __init__(self, m1, m2, m3):
        self.avg = (m1 + m2 + m3) / 3

    def grade(self):
        if self.avg >= 90: return "A"
        elif self.avg >= 75: return "B"
        elif self.avg >= 60: return "C"
        elif self.avg >= 40: return "D"
        else: return "F"

s = Student(80, 85, 90)
print("Grade:", s.grade())
```
